```python
import matplotlib.pyplot as plt

# Epochs
epochs = list(range(8))

# Training and Validation Loss values
train_loss = [0.00001, 0.00001, 0.00000, 0.00000, 0.00001, 0.00000, 0.00000, 0.00000]
val_loss = [0.03698, 0.01247, 0.28396, 0.01115, 0.00742, 0.00381, 0.00363, 0.00382]

# Plotting
# Adjust the figure size here: (width, height)
plt.figure(figsize=(6, 6))  # Adjusted width and height

# Adjust the line color and width
plt.plot(epochs, train_loss, label='Train Loss', marker='o', color='darkorange', linewidth=2)
plt.plot(epochs, val_loss, label='Validation Loss', marker='s', color='dodgerblue', linewidth=2)

plt.title('Training and Validation Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Adjusting the style of the plots to add top and right borders for a "boxed" look
train_accuracy = [0.9838, 0.9942, 0.9963, 0.9976, 0.9961, 0.9999, 1, 1]
valid_accuracy = [0.989, 0.9948, 0.924, 0.9978, 0.9976, 0.9991, 0.9991, 0.9993]
# Set up the figure and axes for the subplots again with a "boxed" look
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plot Training and Validation Loss with adjustments
axes[0].plot(train_loss, label='Train Loss', marker='o', color='darkorange', linewidth=2)
axes[0].plot(val_loss, label='Validation Loss', marker='s', color='dodgerblue', linewidth=2)
#axes[0].set_title('Training and Validation Loss per Epoch')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].grid(True)
axes[0].legend()
# Adding top and right borders
axes[0].spines['top'].set_visible(True)
axes[0].spines['right'].set_visible(True)

# Plot Training and Validation Accuracy with adjustments
axes[1].plot(train_accuracy, label='Train Accuracy', marker='o', color='darkorange', linewidth=2)
axes[1].plot(valid_accuracy, label='Validation Accuracy', marker='s', color='dodgerblue', linewidth=2)
#axes[1].set_title('Training and Validation Accuracy per Epoch')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy')
axes[1].grid(True)
axes[1].set_ylim(0.9, 1.01)  # Adjust y-axis limits
axes[1].legend()
# Adding top and right borders
axes[1].spines['top'].set_visible(True)
axes[1].spines['right'].set_visible(True)

plt.tight_layout()
plt.show()
```

```python
import os
```

```python
def count_image_files(folder_path):
    # List all files in the folder
    files = os.listdir(folder_path)
    
    # Filter for image files (e.g., .jpg, .png, .jpeg, .bmp, .gif)
    image_files = [file for file in files if file.lower().endswith(('.jpg', '.png', '.jpeg', '.bmp', '.gif','.txt'))]
    
    # Count the number of image files
    return len(image_files)
```

```python
# Replace 'your_folder_path' with the path to your folder
folder_path = '../Data/genc/extracted/img'
num_images = count_image_files(folder_path)
print(f"Number of image files in the folder: {num_images}")
```

```python
import os
import matplotlib.pyplot as plt

# Path to the directories
label_dir = '../Data/genc/coco/labels/train2017'  # Replace with your label directory path

# Initialize a dictionary to hold the count of each class
class_counts = {i: 0 for i in range(80)}  # 80 classes as per your list

# Read and process each label file
for filename in os.listdir(label_dir):
    if filename.endswith('.txt'):
        with open(os.path.join(label_dir, filename), 'r') as file:
            for line in file:
                class_id = int(line.split()[0])
                class_counts[class_id] += 1

# Prepare data for plotting
classes = list(class_counts.keys())
counts = list(class_counts.values())

# Plotting
plt.figure(figsize=(15, 10))
plt.bar(classes, counts, color='skyblue')
plt.xlabel('Class ID')
plt.ylabel('Count')
plt.title('Frequency of Each Class in COCO Dataset')
plt.xticks(classes, rotation='vertical')
plt.grid(axis='y')
plt.show()
```
